```python
## Basic Packages

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from pandas.tools.plotting import parallel_coordinates


## Sklearn packages
from sklearn.decomposition import PCA as sklearnPCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.manifold import TSNE

## Notebook display settings
pd.options.display.max_columns = 999
```

```python
## Audio feature descriptions

#https://developer.spotify.com/documentation/web-api/reference/tracks/get-audio-features/
```

```python
## Read in data

combined_df = pd.read_csv('data/combined_df.csv')
```

```python
combined_df.head()
```

```python
## Trimming down the audio features that I want to explore initially

features_df = combined_df[['release_date', 'album_name', 'name', 'acousticness', 'danceability', 'energy', 
    'loudness', 'key', 'mode', 'tempo', 'valence']]
```

```python
## Removing duplicate and greatest hits album

album_drop = ["""1 (Remastered)""", """Sgt. Pepper's Lonely Hearts Club Band (Deluxe Edition)""", """Live At The Hollywood Bowl"""]

features_df = features_df[~features_df['album_name'].isin(album_drop)]
```

```python
## Separating the release date into year, month, day columns

features_df['year'], features_df['month'], features_df['day'] = features_df['release_date'].str.split('-',2).str

## Sort oldest album to newest album

features_df = features_df.sort_values(by = ['year', 'month', 'day'])

## Reset index and assign unique release dates to a list that will be used in the next loop

features_df.reset_index(drop = True, inplace = True)

unique_release = list(features_df.release_date.unique())
```

```python
## For loop that assigns an album number based on the release date for future plotting

album_number = []

count = 0
record_number = 1

for row in range(len(features_df)):
    if features_df.release_date.iloc[row] == unique_release[count]:
        print(features_df.release_date.iloc[row], "-------->", unique_release[count], "------->", features_df.release_date.iloc[row] == unique_release[count])
        album_number.append(record_number)
    else:
        print('Change Album')
        record_number += 1
        count += 1
        album_number.append(record_number)
```

```python
## Converting resulting list into a dataframe to concat, then renaming the column to show album_number

album_number = pd.DataFrame(album_number)

features_df = pd.concat([features_df, album_number], axis = 1)

features_df.rename(columns = {0 : 'album_number', 'name' : 'song_name'}, inplace = True)
```

## Regression plot of individual audio features

```python
reg_features = features_df[['album_number', 'song_name', 'acousticness', 'danceability', 'energy', 'loudness']]
```

```python
reg_features.head()
```

```python
def feature_norm(df, col):
    return(df[col] - df[col].min())/(df[col].max() - df[col].min())
```

```python
## Normalizing the values for each measure

reg_features['acousticness_norm'] = (reg_features.acousticness - reg_features.acousticness.min())/(reg_features.acousticness.max() - reg_features.acousticness.min())
reg_features['danceability_norm'] = (reg_features.danceability - reg_features.danceability.min())/(reg_features.danceability.max() - reg_features.danceability.min())
reg_features['energy_norm'] = (reg_features.energy - reg_features.energy.min())/(reg_features.energy.max() - reg_features.energy.min())
reg_features['loudness_norm'] = (reg_features.loudness - reg_features.loudness.min())/(reg_features.loudness.max() - reg_features.loudness.min())
```

```python
sns.set(rc={'figure.figsize':(15,15)})

_ = sns.pairplot(reg_features.iloc[:,0:6], hue = 'album_number')
```

## PCA Analysis using sklearn

```python
## Creating an album list for labeling

album_list = list(features_df.album_name.unique())
print(album_list)
```

```python
## Reducing features

pca_df = features_df[['album_number', 'acousticness', 'danceability', 'energy', 'loudness', 'key', 'mode', 'tempo', 'valence']]

## Splitting data into target and features

y = pca_df['album_number']
X = pca_df.iloc[:,1:]
```

```python
## Normalizing data

X_norm = (X - X.min())/(X.max() - X.min())

## Defining 4 components for PCA and applying fit_transform

pca = sklearnPCA(n_components = 4)
transformed = pd.DataFrame(pca.fit_transform(X_norm))
```

```python
## Plotting results of PCA transformation

sns.set(rc={'figure.figsize':(12,12)})

for i in range(len(album_list)):
    plt.scatter(transformed[y == i][0], transformed[y == i][1], label = album_list[i], cmap = 'RdYlGn')

plt.legend(bbox_to_anchor=(1, 0.5))
plt.show()
```

## Linear Discriminant Analysis

```python
## Defining 4 components for LDA and applying fit_transform

lda = LinearDiscriminantAnalysis(n_components = 4)
transformed_2 = pd.DataFrame(lda.fit_transform(X_norm, y))
```

```python
## Plotting results of LDA transformation

for i in range(len(album_list)):
    plt.scatter(transformed_2[y == i][0], transformed_2[y == i][1], label = album_list[i], cmap = 'RdYlGn')

plt.legend(bbox_to_anchor=(1, 0.5))
plt.show()
```

## t-SNE clustering

```python
## Defining 4 components for LDA and applying fit_transform

tsne = TSNE(n_components = 2, perplexity = 45, learning_rate = 550)
transformed_3 = pd.DataFrame(tsne.fit_transform(X_norm))
```

```python
## Plotting results of t-SNE transformation

for i in range(len(album_list)):
    plt.scatter(transformed_3[y == i][0], transformed_3[y == i][1], label = album_list[i], cmap = 'RdYlGn')

plt.legend(bbox_to_anchor=(1, 0.5))
plt.show()
```

## Parallel Coordinates

```python
par_cord_df = features_df[['album_number', 'acousticness', 'danceability', 'energy', 'loudness', 'key', 'mode', 'tempo', 'valence']]
```

```python
y_par = par_cord_df['album_number']
X_par = par_cord_df.iloc[:,1:]

## Normalizing data

X_norm_par = (X_par - X_par.min())/(X_par.max() - X_par.min())

# Concat classes with the normalized data
data_norm = pd.concat([X_norm_par[par_cord_df], y], axis=1)

# Perform parallel coordinate plot
parallel_coordinates(data_norm, 'album_name')
plt.show()
```
